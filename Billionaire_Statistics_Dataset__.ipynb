```python
import pandas as pd
```

```python
import io
```

```python
from google.colab import files
```

```python
uploaded_billionaire  = files.upload()
```

```python
df_billionaire = pd.read_csv(io.BytesIO(uploaded_billionaire['Billionaires Statistics Dataset.csv']))
df_billionaire.head(10)
```

```python
# Remove the date column from the DataFrame
df_billionaire_f = df_billionaire.drop('date', axis=1)
```

```python
for col_bill in df_billionaire_f.columns:
   print(col_bill)
```

Droping the birthdate field as there are already 3 features as birth year, birth month, and birthdate

```python
# Remove the birthDate column from the DataFrame
df_billionaire_ff = df_billionaire_f.drop('birthDate', axis=1)
```

```python
for col_bill_ in df_billionaire_ff.columns:
   print(col_bill_)
```

```python
df_billionaire_ff['category'].equals(df_billionaire_ff['industries'])
```

Since category and indutriies column have exactly the same value and no new feature, therefore deleting the category column from the dataset.

```python
# Remove the birthDate column from the DataFrame
df_billionaire_ff__ = df_billionaire_ff.drop('category', axis=1)
```

```python
for col_bill_ in df_billionaire_ff__.columns:
   print(col_bill_)
```

```python
df_billionaire_ff__.to_csv('billionaire.csv', sep=',', index=False, encoding='utf-8')
```
